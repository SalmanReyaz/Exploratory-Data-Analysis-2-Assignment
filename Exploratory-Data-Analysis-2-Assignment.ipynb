# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

 # Load the dataset
df = pd.read_excel('flight_price.xlsx')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Load the dataset
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head(3)
```

```python
#  Feature Selection (Assuming you have columns like 'Date', 'Price', 'Airline', 'Destination', etc.)
date_column = 'Date_of_Journey'
price_column = 'Price'
airline_column = 'Airline'
destination_column = 'Destination'


# Time-Series Analysis
# Ensure the 'Date' column is in datetime format
df[date_column] = pd.to_datetime(df[date_column])
```

```python
df.head()
```

```python
# Bar Chart - Average Prices by Airline
plt.figure(figsize=(10, 6))
sns.barplot(x=airline_column, y=price_column, data=df)
plt.title('Average Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
dataset.head(3)
```

```python
dataset['Price'].unique()
```

```python
dataset
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,100))


# Create a bar plot
plt.bar(x='App',y='Price',data=dataset)

# Add labels and title
plt.xlabel('App Type')
plt.ylabel('Average Price (USD)')
plt.title('Average App Prices by Type')

# Show the plot
plt.show()
```

```python
# Remove rows with non-numeric 'Price' values
dataset = dataset[dataset['Price'].apply(lambda x: x.replace('$', '').replace('Everyone', '').replace('Free', '0').replace('Paid', '').isnumeric())]

# Convert the 'Price' column to a numeric data type
dataset['Price'] = dataset['Price'].apply(lambda x: float(x.replace('$', '')))

```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Group the data by 'Type' (Free or Paid) and calculate the average price for each group
average_prices = dataset.groupby('Type')['Price'].mean().reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_prices['Type'], average_prices['Price'], color='skyblue')
plt.xlabel('App Type')
plt.ylabel('Average Price (USD)')
plt.title('Average Prices by App Type (Google Play)')
plt.xticks(rotation=0)  # Ensure app types are displayed horizontally

# Display the chart
plt.show()
```
